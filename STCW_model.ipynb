```python
import nltk
import re
from nltk.corpus import wordnet as wn
from itertools import product

def get_sim_score(word_1, word_2):
    """ 
    Calculate the highest path similarity among all pairs. 
    """
    if word_1 == word_2:
        return 1
    else:
        max_sim = -1.0
        synsets_1 = wn.synsets(word_1)
        synsets_2 = wn.synsets(word_2)
        if synsets_1 and synsets_2:
            for synset_1, synset_2 in product(synsets_1, synsets_2):
                sim = wn.path_similarity(synset_1, synset_2)
                try:
                    if sim > max_sim:
                        max_sim = sim
                except TypeError:
                    continue
            return max_sim
        return max_sim

def sort_text(origin_list, new_list):
    """
    
    :param ori_list: original list in the original order, must contain every word in the new list
    :param new_list: the new list to be sorted according to the original order
    :return: sorted new_list
    """
    pass
    
    


def mySim(text1, text2, sigma=0.85):
   
    # set stop words 
    stopwords = nltk.corpus.stopwords.words('english')
    x = []
    y = []
    dic = {}
    
    # clean raw text
    text1 = re.sub('[^a-zA-Z]', ' ', text1).lower()
    text2 = re.sub('[^a-zA-Z]', ' ', text2).lower()
    
    # tokenize inputs into vectors 
    token_p = nltk.word_tokenize(text1, language='english')
    token_r = nltk.word_tokenize(text2, language='english')
    
    concept_p = [words for words in token_p if words not in stopwords]
    concept_r = [words for words in token_r if words not in stopwords]
    
    for w1, w2 in product(concept_p, concept_r):
        sim = get_sim_score(w1, w2)
        if sim >= sigma:
            x.append(w1)
            y.append(w2)
            dic[w1] = w2
    
    concept_x = [words for words in token_p if words in x]
    concept_y = [words for words in token_r if words in y]
    
    count = len(concept_x)
    sim_score_0 = 0
    total = 0
    
    if count == 0:
        sim_score_0 = 0
    
    elif count%2 == 0 or count == 1:
        for position, word in enumerate(concept_x):
            total += abs(position - concept_y.index(dic[word]))
        
        sim_score_0 = 1 - 2 * total / count ** 2
            
    else:
        for position, word in enumerate(concept_x):
            total += abs(position - concept_y.index(dic[word]))
        
        sim_score_0 = 1 - 2 * total / (count ** 2 - 1)
    
    
    concept_pp = [words for words in token_p if words not in x]
    concept_rr = [words for words in token_r if words not in y]
    
    for w1, w2 in product(concept_pp, concept_rr):
        print(w1, w2)
    
    print(concept_pp)
    print(concept_rr)
```
